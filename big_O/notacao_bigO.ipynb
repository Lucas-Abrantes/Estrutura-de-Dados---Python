#### Notação BigO

. Busca relacionar diferentes algoritmos;

. Analisar a complexidade envoldida.


```python
# Função O(n)


def soma(n):
    soma = 0
    for i in range(n+1):
        soma += 1
    return soma
```

```python
%timeit soma(10)
```

```python
def soma(n):
    return (n*(n + 1))/2
```

```python
%timeit soma(10)
```

```python
# Função O^2(n)


def lista():
    lista = []
    for i in range(1000):
        lista += [i]
    return lista

print(lista())
```

```python
%timeit lista()
```

```python
def lista2():
    return range(1000)

l = lista2()
for i in (l):
    print(i)
```

```python
%timeit lista2()
```

### Gráfico de uma constante

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.linspace(1, 10, 100)
len(n)
n
```

```python
labels = ['constante', 'logaritmo', 'linear', 'log linear', 'quadratica', 'cubica', 'exponencial']

big_o = [np.ones(n.shape), np.log(n), n, np.log(n)* n, n**2, n**3, 2**n]
```

```python
plt.figure(figsize=(8, 6))
plt.ylim(0, 100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
plt.legend()
plt.ylabel("Runtime")
plt.xlabel("n")
```
